## What is the difference in spending patterns between male and female customers? Do they tend to buy different types of products? What is the relationship between gender and payment type?

```python
import pandas as pd
import os
import csv
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import linregress
import seaborn as sns
```

```python
shopping_behaviors_csv = pd.read_csv('supermarket_sales.csv')
shopping_behaviors_csv
```

```python
#bar chart displaying product line purchases by gender
plt.figure(figsize=(20,5))

colors = ['pink', 'blue']

sns.countplot(x ='Product line',hue='Gender' ,data = shopping_behaviors_csv, palette=colors)
plt.title('Product Category Purchases by Gender')
plt.xlabel('Gender')
plt.ylabel('Product Category')

plt.show()
#data source: Kaggle.com
```

```python
#bar chart displaying total spent by gender
total_spend_by_gender = shopping_behaviors_csv.groupby('Gender')['Total'].sum()

colors = ['pink', 'blue']

plot = total_spend_by_gender.plot(kind='bar', color=colors)

plt.title('Total Cost by Gender')
plt.xticks(rotation=360)
plt.xlabel('Gender')
plt.ylabel('Total Cost')


plt.show()
```

```python
#bar chart displaying total spend per product category per gender
total_spend_by_line = shopping_behaviors_csv.groupby(['Gender', 'Product line'])['Total'].sum().reset_index()

colors = ['pink', 'blue']

plot = total_spend_by_line.pivot(index='Product line', columns='Gender', values='Total').plot(kind='bar', color=colors, figsize=(20,5))

plt.title('Total Spend by Gender and Product Category')
plt.xticks(rotation=360)
plt.xlabel('Product Category')
plt.ylabel('Total Spend')

plt.show()
```

```python
#bar chart displaying total spend per product category per gender
total_spend_by_line_gender = shopping_behaviors_csv.groupby(['Gender', 'Payment'])['CustomerID'].nunique().reset_index()

colors = ['pink', 'blue']

plot_data = total_spend_by_line_gender.pivot(index='Payment', columns='Gender', values='CustomerID')

plot = plot_data.plot(kind='bar', color=colors, figsize=(20,5))

plt.title('Number of Customers by Payment Type and Gender')
plt.xlabel('Payment Type')
plt.ylabel('Number of Customers')
plt.xticks(rotation=360)

plt.show()
```
